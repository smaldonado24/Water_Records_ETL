```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
#Creating connection to Postgres
URL = 'postgresql://postgres:Lakers24@localhost:5432/Snake_Valley'

engine = create_engine(URL)
```

```python
#Pulling a single table from our Postgres
Stations = 'SELECT * FROM "Stations"'
Stations_df = pd.read_sql(Stations, engine)
Stations_df
```

```python
#Pulling from multiple tables from our Postgres
Station_location = '''
SELECT s."Station Name", s."Active", l."Latitude", l."Longitude" 
FROM "Stations" as s 
JOIN "Locations" as l ON l."Station ID" = s."Station ID" 
WHERE s."Station Type" = 'Spring'
'''
Station_Location_df = pd.read_sql(Station_location, engine)
Station_Location_df
```
