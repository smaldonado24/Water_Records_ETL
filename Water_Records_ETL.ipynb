```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fetch the webpage using requests
urls =[ 
    'https://waterrights.utah.gov/cgi-bin/dvrtview.exe?Modinfo=StationView&STATION_ID=9569&RECORD_YEAR=2024',
    'https://waterrights.utah.gov/cgi-bin/dvrtview.exe?Modinfo=StationView&STATION_ID=9572&RECORD_YEAR=2024',
    'https://waterrights.utah.gov/cgi-bin/dvrtview.exe?Modinfo=StationView&STATION_ID=9570&RECORD_YEAR=2024',
    'https://waterrights.utah.gov/cgi-bin/dvrtview.exe?Modinfo=StationView&STATION_ID=9595&RECORD_YEAR=2024',
    'https://waterrights.utah.gov/cgi-bin/dvrtview.exe?Modinfo=StationView&STATION_ID=9573&RECORD_YEAR=2024',
    'https://waterrights.utah.gov/cgi-bin/dvrtview.exe?Modinfo=StationView&STATION_ID=10318&RECORD_YEAR=2024',
    'https://waterrights.utah.gov/cgi-bin/dvrtview.exe?Modinfo=StationView&STATION_ID=9600&RECORD_YEAR=2024',
    'https://waterrights.utah.gov/cgi-bin/dvrtview.exe?Modinfo=StationView&STATION_ID=9574&RECORD_YEAR=2024',
    'https://waterrights.utah.gov/cgi-bin/dvrtview.exe?Modinfo=StationView&STATION_ID=9599&RECORD_YEAR=2024',
    'https://waterrights.utah.gov/cgi-bin/dvrtview.exe?Modinfo=StationView&STATION_ID=9589&RECORD_YEAR=2024',
    'https://waterrights.utah.gov/cgi-bin/dvrtview.exe?Modinfo=StationView&STATION_ID=9571&RECORD_YEAR=2024'
]

def extract_data(url):
    response = requests.get(url)
    data = {
            'System Name': None,
            'Station Name': None,
            'Description': None,
            'System Group': None,
            'System Subgroup': None,
            'Status': None,
            'Diverting Works': None,
            'Measuring Device': None,
            'Location': None,
            'Record Type': None,
            'Record Rating': None,
            'Record Period': None
        }

# Check if the request was successful (status code 200)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        pre_tag = soup.find('pre')

    # If the <pre> tag exists, extract its text and process it
        if pre_tag:
            text = pre_tag.get_text()

        # Split the text by newlines and process line by line
            for line in text.splitlines():
                if 'System Name' in line:
                    data['System Name'] = line.split(':')[1].strip()
                if 'Station Name' in line:
                    data['Station Name'] = line.split(':')[1].strip()
                if 'Description' in line:
                    data['Description'] = line.split(':')[1].strip()
                if 'System Group' in line:
                    data['System Group'] = line.split(':')[1].strip()
                if 'System Subgroup' in line:
                    data['System Subgroup'] = line.split(':')[1].strip()
                if 'Status' in line:
                    data['Status'] = line.split(':')[1].strip()
                if 'Diverting Works' in line:
                    data['Diverting Works'] = line.split(':')[1].strip()
                if 'Measuring Device' in line:
                    data['Measuring Device'] = line.split(':')[1].strip()
                if 'Location' in line:
                    data['Location'] = line.split(':')[1].strip()
                if 'Record Type' in line:
                    data['Record Type'] = line.split(':')[1].strip()
                if 'Record Rating' in line:
                    data['Record Rating'] = line.split(':')[1].strip()
                if 'Record Period' in line:
                    data['Record Period'] = line.split(':')[1].strip()

    return data
        
all_data = []

for url in urls:
    station_data = extract_data(url)
    all_data.append(station_data)

stations_df = pd.DataFrame(all_data)

stations_df
```

```python
stations_df_drop = stations_df.drop(columns=["System Group", "System Subgroup", "Diverting Works", "Location"])
stations_df_drop
```

```python
stations_df_drop['Station Type'] = ['Spring', 'Spring', 'Stream', 'Spring', 'Spring', 'Well', 'Spring', 'Spring', 'Spring','Spring', 'Stream']
stations_df_drop
```
